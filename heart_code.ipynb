# Features
<br>
Feature 1 - Age: numerical data that indicates the individuals age <br>
Feature 2 - Sex: categorical data either being M for male or F for female<br>
Feature 3 - ChestPain: categorical data with 4 types, ASY meaning no chest pain, NAP meaning chest pain at night, ATA meaning nontypical angina, TA meaning typical angina<br>
Feature 4 - RestingBp: numerical data representing the indivduals blood pressure when they're in a rest state<br>
Feature 5 - Cholesterol: numerical data of the individuals cholesterol level<br>
Feature 6 - Fasting blood sugar: Numerical data of the individuals blood sugar level when they haven't eaten for atleast 8 hours<br>
Feature 7 - RestingECG: Categorical data which ecords the electrical activity of your heart while you are at rest, there is 3 types, normal being a stable heartbeats per minute, LVH being  heart's left ventricle thickens, making it harder for the heart to pump blood, and ST being an indication for injury to the heart muscle, or myocardium<br>
Feature 8 - MaxHR: Numerical data indicating number of times an individual heartbeats per minute during intense physical activity<br>
Feature 9 - ExerciseAngina: Categorical data, Y meaning true that exercise aggragates angina, and N meaning false that exercise doesn't aggragate angina<br>
Feature 10 - Oldpeak: Numerical data, a measurement of the depression in an electrocardiogram (ECG) between the first and second parts of the ST segment<br>
Feature 11 - ST_Slope: Categorical data, the slope of the peak exercise ST segment, two major types being flat so stable and up having a upwards trend slope and down having a downward slope<br>
Feature 12 - HeartDisease - Numerical data indicating whether the individual has heart disease or not, 0 meaning no, 1 meaning yes<br>

# Insight 1 - (r1) Middle aged individuals tend to have cases of heart disease more compared to younger individuals and much older individuals

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart.csv'  # Replace with the actual path to your heart.csv file
data = pd.read_csv(file_path)

# Plot a histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=10, kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

We can see from this graph that ages 50 to 60 have the most cases of heart disease. There might be fewer individuals with heart disease in very old age groups. This could reflect survival bias, where older individuals with heart disease may no longer be represented in the dataset due to higher mortality rates past 60 years old.

# Insight 3 - Grouping the upward and downward slope and comparing the count to the flat slope

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart.csv'  # Replace with the actual path to your heart.csv file
data = pd.read_csv(file_path)

# Create a new column 'ST_Slope_Group' to group 'Up' and 'Down' into one category, and 'Flat' into another
data['ST_Slope_Group'] = data['ST_Slope'].apply(lambda x: 'Up/Down' if x in ['Up', 'Down'] else 'Flat')

# Create a count plot with the grouped 'ST_Slope_Group'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='ST_Slope_Group', palette='Set2')

# Customize the plot
plt.title('Frequency Distribution of Grouped ST_Slope (Up/Down vs Flat)', fontsize=14)
plt.xlabel('ST_Slope Group', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

We can see they are about equal with the flat slope just being a little bit above so there is no direct correlation between the slope trend in the ST segment

# Insight 4 - (r4) Patients with ExerciseAngina = Yes (Y) have a significantly higher likelihood of heart disease compared to those with ExerciseAngina = No (N)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


file_path = "heart.csv"  
df = pd.read_csv(file_path)

# Calculate percentage of heart disease within each ExerciseAngina group
angina_heart_disease = (
    df.groupby("ExerciseAngina")["HeartDisease"]
    .mean()
    .reset_index()
    .rename(columns={"HeartDisease": "HeartDiseaseRate"})
)

# Plotting the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=angina_heart_disease, x="ExerciseAngina", y="HeartDiseaseRate", palette="coolwarm")
plt.title("Heart Disease Rate by Exercise Angina")
plt.xlabel("Exercise Angina (Y/N)")
plt.ylabel("Percentage with Heart Disease")
plt.ylim(0, 1)
plt.show()
```

The graph clearly shows that ExerciseAngina = Yes (Y) is strongly associated with an increased rate of heart disease

# Insight 5 - (r5) - RestingECG = ST is associated with a significantly higher risk of heart disease compared to other categories

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from heart.csv
file_path = "heart.csv"  # Ensure the file is in the same directory or provide the correct path
df = pd.read_csv(file_path)

# Calculate heart disease rate for each RestingECG category
restingecg_heart_disease = (
    df.groupby("RestingECG")["HeartDisease"]
    .mean()
    .reset_index()
    .rename(columns={"HeartDisease": "HeartDiseaseRate"})
)

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(
    data=restingecg_heart_disease,
    x="RestingECG",
    y="HeartDiseaseRate",
    palette="plasma"
)
plt.title("Heart Disease Rate by RestingECG")
plt.xlabel("Resting ECG Category")
plt.ylabel("Percentage with Heart Disease")
plt.ylim(0, 1)
plt.show()
```

As shown in the graph we can see a larger margin for ST with individuals that have heart disease

# Insight 6 - (r6) There is a negative correlation between Age and MaxHR (Maximum Heart Rate achieved), meaning as age increases, the maximum heart rate achieved tends to decrease.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the data from heart.csv
file_path = "heart.csv"  # Ensure the file is in the same directory or provide the correct path
df = pd.read_csv(file_path)

# Scatterplot: Age vs MaxHR
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Age", y="MaxHR", alpha=0.7, color="blue")
plt.title("Scatterplot of Age vs MaxHR")
plt.xlabel("Age")
plt.ylabel("MaxHR (Maximum Heart Rate)")
plt.grid(True)
plt.show()

# Calculate Pearson correlation coefficient
correlation, p_value = pearsonr(df["Age"], df["MaxHR"])
print(f"Pearson Correlation Coefficient: {correlation:.2f}")
```

The scatterplot shows a downward trend indicating a negative correlation

# Insight 8 - (r6) There is a positive correlation between Cholesterol and Oldpeak (ST depression induced by exercise relative to rest

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the data from heart.csv
file_path = "heart.csv"  # Ensure the file is in the same directory or provide the correct path
df = pd.read_csv(file_path)

# Scatterplot: Cholesterol vs Oldpeak
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Cholesterol", y="Oldpeak", alpha=0.7, color="green")
plt.title("Scatterplot of Cholesterol vs Oldpeak")
plt.xlabel("Cholesterol (mg/dL)")
plt.ylabel("Oldpeak (ST Depression)")
plt.grid(True)
plt.show()

# Calculate Pearson correlation coefficient
correlation, p_value = pearsonr(df["Cholesterol"], df["Oldpeak"])
print(f"Pearson Correlation Coefficient: {correlation:.2f}")
```

As cholesterol levels increase, there is a slight tendency for higher ST depression during exercise, indicating possible worsening cardiac stress
