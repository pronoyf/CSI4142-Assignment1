## Loading the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt




# Load the dataset
file_path = "https://raw.githubusercontent.com/pronoyf/CSI4142-Assignment1/refs/heads/main/shootings.csv"
dataset_2 = pd.read_csv(file_path)
```

### Insight 1(r1)

The prime majority of the individuals who have been shot are in the 20-40 range, where there are fewer shooting incidents the older you go. According to the age distribution historogram below, the majority of the victims of these police shootings fall within that age range and subsequently the frequency of shootings past that age lessens the older the graph goes. This was done using a univariate analysis of the age attribute, which is numerical. It uses a historogram to display a distribution of the age feature. The code will be provided below to produce the graph always.

```python

plt.figure(figsize=(8, 5))
plt.hist(dataset_2['age'].dropna(), bins=20, edgecolor="black", alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age Distribution in Police Shootings")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Insight 2 (r2)

White, Black, and Hispanic are the majority races of the indidiuals who are victims to police shootings in the US. The countplot below shows the racial distribution of victims in police shootings. The White, Black, and Hispanic categories have the highest counts, while Asian, Native American, and Other racial groups have significantly lower numbers. This is a univariate analysis(r2) of the categorical feature 'race', using a countplot to illustrate the total counts of each category in the 'race' feature.

```python
import seaborn as sns


plt.figure(figsize=(8, 5))
sns.countplot(data=dataset_2, x="race", order=dataset_2["race"].value_counts().index, palette="viridis")
plt.xlabel("Race")
plt.ylabel("Count")
plt.title("Count of Police Shootings by Race")
plt.xticks(rotation=45)
plt.show()
```

### Insight 3(r3)

The South and West regions have the highest number of police shootings and so perhaps there tend to be more police indients in those regions due to higher crime or other policies. As displayed by the countplot below, there is evidently a higher count in the South and West groups. The countplot below categorizes police shootings based on U.S. regions:South and West account for the majority of cases; Midwest has a moderate number of incidents; Northeast has the lowest frequency of police shootings. This is a Univariate Analysis of Categorical Data (r3) using a Grouped Countplot, where states were classified into major U.S. regions.

```python
region_mapping = {
    "Northeast": ["CT", "ME", "MA", "NH", "RI", "VT", "NJ", "NY", "PA"],
    "Midwest": ["IL", "IN", "MI", "OH", "WI", "IA", "KS", "MN", "MO", "NE", "ND", "SD"],
    "South": ["DE", "FL", "GA", "MD", "NC", "SC", "VA", "WV", "AL", "KY", "MS", "TN", "AR", "LA", "OK", "TX"],
    "West": ["AZ", "CO", "ID", "MT", "NV", "NM", "UT", "WY", "AK", "CA", "HI", "OR", "WA"]
}

def map_region(state):
    for region, states in region_mapping.items():
        if state in states:
            return region
    return "Unknown"
```

```python

dataset_2["region"] = dataset_2["state"].apply(map_region)
plt.figure(figsize=(8, 5))
sns.countplot(data=dataset_2, x="region", order=dataset_2["region"].value_counts().index, palette="pastel")
plt.xlabel("U.S. Region")
plt.ylabel("Count")
plt.title("Police Shootings by U.S. Region")
plt.show()
```

In the countplot below we compare the count of killings where the metrics are the gender of the victim, as well as the manner of their death. As shown, the number of males shot is incredibly high as opposed to women. ### Insight 6(r6)

The year 2020 saw a noticeable dip in fatal police shootings, lower than surrounding years. The scatterplot now highlights the 2020 drop.
Compared to other years, 2020 had a lower number of fatal police shootings;
This downward spike could be related to the COVID-19 lockdowns, reduced public interactions. It is also possible police policies may have shifted, although this may also be influenced by COVID-19 policies as well. This is meant to use a bivariate analysis(r6). However, due to the fact that this dataset lacks a second numerical feature for a scatterplot, we have chosen to use year as the first numerical attribute and total count of shootings per year as the second metric.At the end we have a scatterplot comparing year (numerical) vs. total shootings per year (numerical count) and there is a noticeable change that marks a correlation of sorts. Besides the 2020 drip there are small fluctuations but the trend seems to be that total shootings have been going downwards with a heavy dip in numbers at 2020. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset_2["date"] = pd.to_datetime(dataset_2["date"], errors='coerce')  
dataset_2["year"] = dataset_2["date"].dt.year  
yearly_counts = dataset_2["year"].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.scatterplot(x=yearly_counts.index, y=yearly_counts.values, color='blue', s=100, alpha=0.7)
plt.plot(yearly_counts.index, yearly_counts.values, linestyle='dashed', color='red')
plt.title("Total Fatal Police Shootings Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Shootings")
plt.show()
```

### Insight 7

The median age of fatal police shooting victims varies by race, with White victims having a slightly higher median age than other racial groups. The boxplot illustrates the distribution of ages across different racial groups. White victims tend to be slightly older on average compared to other racial groups. Other racial groups, such as Black and Hispanic individuals, have a lower median age but a similar overall distribution. ### Insight 8 

Only about 11% of fatal police shootings were recorded on body cameras, meaning the nature of these encounters does not have video proof of what happened. The bar chart shows that body cameras were used in only 11.8% of cases, meaning nearly 88.2% of fatal police shootings were not recorded.
This raises concerns about transparency and accountability of the cases that were not recorded. This is a univariate analysis(r2) of the categorical data body_camera where we calculte the percentage and display that instead of the raw numbers.

```python
body_cam_percentage = dataset_2["body_camera"].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 5))
sns.barplot(x=body_cam_percentage.index, y=body_cam_percentage.values, palette="coolwarm")
plt.title("Percentage of Fatal Police Shootings Recorded on Body Cameras")
plt.xlabel("Body Camera Used")
plt.ylabel("Percentage")
plt.xticks([0, 1], ["No", "Yes"])
plt.show()
```

### Insight 9 



Cases classified as "attack" are the most frequent, and body camera usage is relatively low across all threat levels. The countplot orders the threat level categories based on frequency. "Attack" is the most frequently assigned threat level, followed by "Other" and "Undetermined." Body camera usage remains relatively low across all threat levels, meaning most fatal police shootings are not recorded regardless of the perceived threat level. This is a bivariate(r5) analysis comparing body_camera and threat_level where threat level is ordered explicitly by high count. We compare two categorical values where threat_level has more than 2 values. 

```python


ordered_threat_levels = dataset_2["threat_level"].value_counts().index
plt.figure(figsize=(8, 5))
sns.countplot(x="threat_level", hue="body_camera", data=dataset_2, order=ordered_threat_levels, palette="Set2")
plt.title("Body Camera Usage by Threat Level in Fatal Police Shootings")
plt.xlabel("Threat Level")
plt.ylabel("Count")
plt.legend(title="Body Camera Used", labels=["No", "Yes"])
plt.xticks(rotation=45)
plt.show()
```

### Insight 10

Body camera usage in fatal police shootings varies significantly by state, with some states using them more frequently than others. As can be seen by the countplot, states like California and Texas have a higher count of shootings but also higher counts of body cameras being used. The countplot compares body camera usage across the top 10 states with the most fatal police shootings. Some states show higher rates of body camera usage, while others have most cases unrecorded. This possibly suggests that state-level policies and regulations influence the adoption of body cameras in police encounters. This is a bivariate analysis (r5), Comparing a categorical variable, body_camera,  across multiple ordered categories (state), sorted by frequency.

```python
ordered_states = dataset_2["state"].value_counts().index[:10]  

plt.figure(figsize=(12, 5))
sns.countplot(x="state", hue="body_camera", data=dataset_2, order=ordered_states, palette="coolwarm")
plt.title("Body Camera Usage in Fatal Police Shootings by State (Top 10)")
plt.xlabel("State")
plt.ylabel("Count")
plt.legend(title="Body Camera Used", labels=["No", "Yes"])
plt.xticks(rotation=45, ha="right")
plt.show()
```
